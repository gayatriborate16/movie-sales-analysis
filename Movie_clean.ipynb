```python
import pandas as pd
import numpy as np
```

```python
movie_df = pd.read_excel("movie.xlsx")
movie_df
```

```python
movie_df['Genre'] = movie_df['Genre'].str.replace(',', ', ', regex=False)
```

```python
genre_cols = movie_df['Genre'].str.get_dummies(sep=', ')
movie_df = pd.concat([movie_df, genre_cols], axis=1)
```

```python
genre_cols.columns.tolist()
```

```python
movie_df
```

```python
movie_df.columns
```

```python
movies_df = movie_df[[
    'Title',
    'Year',
    'Rating',
    'Votes',
    'Runtime (Minutes)',
    'Revenue (Millions)',
    'Metascore',
    'Director'
]].copy()
```

```python
movies_df.to_csv('movies_main.csv', index=False)
```

```python
genre_df = movie_df[['Title', 'Genre', 'Rating', 'Revenue (Millions)', 'Metascore']].copy()
genre_df['Genre'] = genre_df['Genre'].str.split(', ')
genre_df = genre_df.explode('Genre')
```

```python
genre_df
```

```python
actors_df = movie_df[['Title', 'Actors']].copy()
actors_df['Actors'] = actors_df['Actors'].str.split(', ')
actors_df = actors_df.explode('Actors')
actors_df
```

```python
genre_df.to_csv('movies_genre.csv', index=False)
```

```python
actors_df.to_csv('movies_actors.csv', index=False)
```
